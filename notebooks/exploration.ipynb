# brazilian e-commerce data analysis: pandas data exploration

```python
import pandas as pd
```

```python
# read data function
def read_data(file_name):
    data = pd.read_csv(f"../data/{file_name}.csv")
    return data
```

```python
# datasets dict
dataset_names = [
    "customers",
    "geolocation",
    "order_items",
    "order_payments",
    "order_reviews",
    "orders",
    "product_category",
    "products",
    "sellers"
]
```

```python
# reading the dataset
datasets = {
    name: read_data(name) for name in dataset_names
}
```
